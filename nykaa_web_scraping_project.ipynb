# Importing Necessary Libraries


```python
from bs4 import BeautifulSoup
import requests
import datetime
import time
import csv
import pandas as pd
import re
```

# Connecting To Website: Nykaa

**Face Wash**

```python
# Pulling category, title, size, original_price, current_price, discount, ratings, ratings count from Nykaa for Simple Face Wash

url_facewash='https://www.nykaa.com/simple-kind-to-skin-refreshing-facial-wash/p/11093299?productId=11093299&pps=1&skuId=390982'
request_headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36'}

page_facewash=requests.get(url_facewash,headers=request_headers)

soup1_facewash=BeautifulSoup(page_facewash.content,'html.parser')
soup2_facewash=BeautifulSoup(soup1_facewash.prettify(),'html.parser')

# Pulling Category
category_facewash=soup2_facewash.find(class_='last-list css-hnjjmz').get_text(strip=True)

# Pulling Title
title_facewash = soup2_facewash.find('h1', class_='css-1gc4x7i').get_text(strip=True)
clean_title_facewash = re.sub(r'\s*\(.*?\)', '', title_facewash).strip()

# Pulling Size
size_ml_facewash=soup2_facewash.find(class_='css-1ctpgu6').get_text(strip=True)
size_match_facewash=re.search(r'\((\d+)',size_ml_facewash)
size_facewash=int(size_match_facewash.group(1)) if size_match_facewash else None

# Pulling Original Price
original_price_rupee_facewash=soup2_facewash.find(class_='css-u05rr').get_text(strip=True)
original_price_facewash=original_price_rupee_facewash [5:]

# Pulling Current Price
current_price_rupee_facewash=soup2_facewash.find(class_='css-1jczs19').get_text(strip=True)
current_price_facewash=current_price_rupee_facewash [1:]

# Pulling Discount
discount_symbol_facewash=soup2_facewash.find(class_='css-bhhehx').get_text(strip=True)
discount_facewash=discount_symbol_facewash [:-5]

# Pulling Ratings
rating_total_facewash=soup2_facewash.find(class_='css-m6n3ou').get_text(strip=True)
rating_facewash=rating_total_facewash [:-2]

# Pulling Ratings Count
rating_count_all_facewash=soup2_facewash.find(class_='css-1hvvm95').get_text(strip=True)
match_facewash=re.search(r'\d+',rating_count_all_facewash)
rating_count_facewash=int(match_facewash.group()) if match_facewash else None

print(category_facewash)
print(clean_title_facewash)
print(size_facewash)
print(original_price_facewash)
print(current_price_facewash)
print(discount_facewash)
print(rating_facewash)
print(rating_count_facewash)
```

```python
# Creating a date variable to track when data was collected into our CSV file

today=datetime.date.today()
print(today)
```

# Converting Into A CSV File

```python
header=['Date','Category','Product','Size(ml)','Original Price(Rs)','Current Price(Rs)','Discount(%)','Rating(out of 5)','Rating Count']
data=[today,category_facewash,clean_title_facewash,size_facewash,original_price_facewash,current_price_facewash,discount_facewash,rating_facewash,rating_count_facewash]

with open('NykaaWebScraper_Dataset.csv', 'w', newline='',encoding='UTF8') as f:
    writer=csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```

```python
# Creating a function to extract data from all the products

for url in product_urls:
    response=requests.get(url,headers=request_headers)
    soup1=BeautifulSoup(response.content,'html.parser')
    soup2=BeautifulSoup(soup1.prettify(),'html.parser')
    
    # Pulling Category
    category=soup2.find(class_='last-list css-hnjjmz').get_text(strip=True)

    # Pulling Title
    title = soup2.find('h1', class_='css-1gc4x7i').get_text(strip=True)
    clean_title = re.sub(r'\s*\(.*?\)', '', title).strip()

    # Pulling Size
    size_ml=soup2.find(class_='css-1ctpgu6').get_text(strip=True)
    size_match=re.search(r'\((\d+)',size_ml)
    size=int(size_match.group(1)) if size_match else None

    # Pulling Original Price
    original_price_rupee=soup2.find(class_='css-u05rr').get_text(strip=True)
    original_price=original_price_rupee [5:]

    # Pulling Current Price
    current_price_rupee=soup2.find(class_='css-1jczs19').get_text(strip=True)
    current_price=current_price_rupee [1:]

    # Pulling Discount
    discount_symbol=soup2.find(class_='css-bhhehx').get_text(strip=True)
    discount=discount_symbol [:-5]

    # Pulling Ratings
    rating_total=soup2.find(class_='css-m6n3ou').get_text(strip=True)
    rating=rating_total [:-2]

    # Pulling Ratings Count
    rating_count_all=soup2.find(class_='css-1hvvm95').get_text(strip=True)
    match=re.search(r'\d+',rating_count_all)
    rating_count=int(match.group()) if match else None

    # Adding Date Column
    today=datetime.date.today()

    products_data=[today,category, clean_title, size, original_price, current_price, discount, rating, rating_count]

    with open('NykaaWebScraper_Dataset.csv', 'a', newline='',encoding='UTF8') as f:
        writer=csv.writer(f)
        writer.writerow(products_data)

```
